## üìö Functions & Operations Overview

### üîπ **Data Manipulation**
- **Groupby**: Groups data based on one or more keys to perform aggregate calculations.
- **Apply**: Applies a function along an axis of the DataFrame (rows or columns).
- **Agg**: Allows multiple aggregation operations (mean, sum, etc.) simultaneously on grouped data.
- **Lambda**: Defines short, inline functions for quick, concise transformations.
- **reset_index()**: Adds or resets a numerical index to the DataFrame.

### üîπ **Encoding Techniques**
- **One-Hot Encoding**: Converts categorical variables into binary indicator (dummy) variables.
- **Label Encoding**: Encodes categorical variables as numerical values to simplify analysis.

### üîπ **Datetime Operations**
- **pd.to_datetime(df['date'])**: Converts a column into datetime data type.
- **pd.DatetimeIndex(df['date']).month**: Extracts the month component from datetime attributes.

### üìä **Visualization**
- **boxplot**: Displays distribution, central tendency, and outliers.
- **histplot**: Shows frequency distributions of numerical data.
- **scatterplot**: Visualizes relationships between two numerical variables.
- **stripplot**: Plots data points for categorical data to show distribution and spread.
- **barplot**: Compares categorical data based on aggregate values.
- **countplot**: Shows counts of observations in categorical bins.
- **FacetGrid**: Creates a grid of plots to visualize subsets of data easily.
- **pairplot**: Visualizes pairwise relationships among multiple numerical variables.

<h2 style='color:#1db6f5'><b> Boxplot:<b></h2>
<h3> A boxplot is a standardized way of displaying the distribution of data based on a five number summary (‚Äúminimum‚Äù, first quartile (Q1), median, third quartile (Q3), and ‚Äúmaximum‚Äù). It can tell you about your outliers and what their values are. </h3>

<h2 style='color:#1db6f5'><b> Kdeplot:<b></h2>
<h3>A kernel density estimate (KDE) plot is a method for visualizing the distribution of observations in a dataset, analagous to a histogram. KDE represents the data using a continuous probability density curve in one or more dimensions.
 </h3>

```python
imp_numerical_cols=['Sales','Profit','Quantity','Discount']
for i,col in enumerate(imp_numerical_cols):
    text="<center><h3><b  style='color:black'>" + f"Distribution of</b> <b style='color:#1db6f5'> {col}"+ "</b></h3><hr></center>"
    display(HTML(text))
    plt.figure(figsize=(16,5))
    bins='auto'
    if i in [0,1]:
        bins=25
    plt.subplot(1,3,1)
    sns.histplot(data=df,x=col,stat='density',bins=bins,color='#a2d9ce')
    plt.title(f'Histogram of {col}',size=15)
    
    plt.subplot(1,3,2)
    sns.kdeplot(data=df,x=col,color='#3498db')
    plt.title(f'kdeplot of {col}',size=15)
    
    plt.subplot(1,3,3)
    sns.boxplot(data=df,x=col,color='#fad7a0')
    plt.title(f'Box-plot of {col}',size=15)
    plt.tight_layout()
    plt.show()
```

# <b id='dist-num-cols'>Detect </b> <b style='color:#1db6f5'> outliers</b>

```python
# Analyze outlier values in the 'Sales' variable

# Calculate the first and third quartiles (Q1 and Q3)
Q1_sales = df['Sales'].quantile(0.25)
Q3_sales = df['Sales'].quantile(0.75)
IQR_sales = Q3_sales - Q1_sales  # Interquartile Range

# Define boundaries to detect outliers
lower_bound_sales = Q1_sales - 1.5 * IQR_sales
upper_bound_sales = Q3_sales + 1.5 * IQR_sales

# Filter the outlier values
outliers_sales = df[(df['Sales'] < lower_bound_sales) | (df['Sales'] > upper_bound_sales)]

# Display a summary of detected outliers
outliers_summary = {
    'Total rows': df.shape[0],
    'Outliers detected': outliers_sales.shape[0],
    'Outlier percentage': round(100 * outliers_sales.shape[0] / df.shape[0], 2),
    'Minimum sales among outliers': outliers_sales['Sales'].min(),
    'Maximum sales among outliers': outliers_sales['Sales'].max()
}

outliers_summary
```

```python
# Add an indicator column for outlier detection
df['is_outlier_sales'] = (df['Sales'] < lower_bound_sales) | (df['Sales'] > upper_bound_sales)

# Filter only the outlier records
outlier_sales_df = df[df['is_outlier_sales']]

# Analyze: classify outliers by category and check if the profit is negative
category_outlier_analysis = (
    outlier_sales_df
    .assign(Negative_Profit = outlier_sales_df['Profit'] < 0)  # Create a boolean column for negative profits
    .groupby('Category')
    .agg(
        Total_Outliers=('Sales', 'count'),  # Count total outliers per category
        Outliers_with_Negative_Profit=('Negative_Profit', 'sum'),  # Count how many have negative profit
        Percentage_Negative_Profit=('Negative_Profit', lambda x: round(100 * x.sum() / x.count(), 2))  # Calculate percentage
    )
    .reset_index()
)

category_outlier_analysis
```

## üìä Insights from Outlier Analysis

- In the **Furniture** category, a significant portion of high-sales transactions resulted in negative profits (**35.12%**).
- In **Office Supplies**, the rate of negative profit among outlier sales is more moderate (**16.33%**).
- In **Technology**, only a small portion (**11.5%**) of high-sales transactions resulted in losses.

### üß† Interpretation:
- High-ticket items in the Furniture category often lead to losses, potentially due to high logistics costs, aggressive discounts, or suboptimal pricing strategies.
- Office Supplies and Technology sectors are relatively more profitable even among large transactions.

```python
# Analyze sub-categories among outlier sales
subcategory_outlier_analysis = (
    outlier_sales_df
    .assign(Negative_Profit = outlier_sales_df['Profit'] < 0)
    .groupby('Sub-Category')
    .agg(
        Total_Outliers=('Sales', 'count'),
        Outliers_with_Negative_Profit=('Negative_Profit', 'sum'),
        Percentage_Negative_Profit=('Negative_Profit', lambda x: round(100 * x.sum() / x.count(), 2)),
        Average_Profit=('Profit', 'mean')
    )
    .sort_values(by='Percentage_Negative_Profit', ascending=False)
    .reset_index()
)

subcategory_outlier_analysis
```
